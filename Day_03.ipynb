```python
categories = ['A','B','C','D']
values = [10,5,7,9]
plt.bar(categories,values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[4,7,1,3,9]
# x,y = data[],data[]
plt.plot(x,y,linestyle = '--',marker = 'o',color = 'red',label = 'Data')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Customized Line Graph')
plt.legend()
plt.show()
```

```python
categories=['A','B','C','D']
value1=[4,7,3,9]
value2=[2,3,6,9]
# x,y = data[],data[]
plt.bar(categories,value1,label='Value 1')
plt.bar(categories,value2,label='Value 2',bottom = value1)
plt.xlabel('Categories')
plt.ylabel ('Values')
plt.title('Stacked Bar Chart')
plt.legend()
plt.show()
```
