```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.decomposition import PCA
# generating random data points
np.random.seed(0)
x = np.random.randn(100,2)
# initialize PCA with the desired number of components
pca = PCA(n_components = 1)
# transformed the data
transformed_x = pca.fit_transform(x)
transformed_x
# inverse transformed
inv_transformed = pca.inverse_transform(transformed_x)
inv_transformed
plt.scatter(x[:,0],x[:,1])
plt.show
plt.scatter(inv_transformed[:,0],inv_transformed[:,1])
plt.show
```

# K_Mean on numerical data

```python
from sklearn.cluster import KMeans
# generating random data points

np.random.seed(0)
x = np.random.rand(100,2)
# no. of clusters
k=4
# kmean model
kmeans = KMeans(n_clusters=k) 
# fitting the k mean model
kmeans.fit(x)
```

```python
# get the cluster centers 
centres = kmeans.cluster_centers_
centres
```

```python
# get the cluster labels
labels = kmeans.labels_
labels
```

```python
plt.scatter(x[:,0],x[:,1], c = labels, cmap = 'viridis',s = 100)
plt.scatter(centres[:,0],centres[:,1],c='black',marker='*',s = 100,label='Cluster Centers')
plt.title('Kmean Clustering')
plt.legend()
plt.show()
```
